```python
cat=[]
num=[]
for i in df1.columns:
    if df1[i].nunique()<10:
        print(i,df1[i].nunique(),df1[i].unique())
        cat.append(i)
    else:
        num.append(i)
```

```python
num
```

```python
 for i in num:
    fig, ax=plt.subplots(1,2)
    ax[0].hist(df1[i],color='r')
    ax[1].boxplot(df1[i])
    plt.suptitle(i)
    plt.show()
```

```python
df1
```

```python
for i in num:
    q1=df1[i].quantile(0.25)
    q3=df1[i].quantile(0.75)
    iqr=q3-q1
    lf=q1-(1.5*iqr)
    uf=q3+(1.5*iqr)
    df1.loc[df1[i]<lf,i]=lf
    df1.loc[df1[i]>uf,i]=uf
```

```python
X_df=df1.drop('avgdeathsperyear',axis=1)
y=df1['avgdeathsperyear']
```

```python
X_df
y
```

```python
X=X_df.values
Y=y.values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr_model=LinearRegression()
lr_model.fit(X_train,y_train)
```

```python
y_train_pred=lr_model.predict(X_train)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
mse_t=mean_squared_error(y_train_pred,y_train)
rmse_t=mse_t**0.5
mae_t=mean_absolute_error(y_train_pred,y_train)
r2_train=r2_score(y_train,y_train_pred)
print(mse_t)
print(rmse_t)
print(mae_t)
print(r2_train)
```

```python
y_test_pred=lr_model.predict(X_test)
```

```python
mse_test=mean_squared_error(y_test_pred,y_test)
rmse_test=mse_test**0.5
mae_test=mean_absolute_error(y_test_pred,y_test)
r2_test=r2_score(y_test,y_test_pred)
print(mse_test)
print(rmse_test)
print(mae_test)
print(r2_test)
```
